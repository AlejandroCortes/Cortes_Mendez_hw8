```python
#calculando el Hessiano
Hessiano=np.zeros([nf-1,nc-1,2,2])
H=np.zeros((2,2))
for i in range(1,(nf-1)):
    for j in range(1,(nc-1)):
        H[0,0]=I[i,j+1]+I[i,j-1]-(2*I[i,j])
        H[1,1]=I[i+1,j]+I[i-1,j]-(2*I[i,j])
        H[0,1]=(I[i+1,j+1]+I[i-1,i-j]+(2*I[i,j])-I[i,j+1]-I[i,j-1]-I[i+1,j]-I[i-1,j])*0.5
        Hessiano[i,j]=H
```

```python
#calculando los eigenvalores y eigenvalores del Hessiano de cada pixel

Eigenvalores=np.zeros([nf-1,nc-1,1,2])
Eigenvectores=np.zeros([nf-1,nc-1,2,2])
for i in range(1,(nf-1)):
    for j in range(1,(nf-1)):
        eigenvalues = eig(Hessiano[i,j])[0]
        eigenvectors= eig(Hessiano[i,j])[1]
        Eigenvalores[i,j]=eigenvalues
        Eigenvectores[i,j]=eigenvectors
```

```python
def circulo_model(nf,nc,x_c,y_c,r_0,r_1,m,phi):
    A=np.zeros([nf,nc])
    for i in range(nf):
        for j in range(nc):
            r=math.sqrt((y_c-i)**2+(x_c-j)**2)
            if(r!=0):
                theta=arcsin((y_c-i)/r)
                R=r+(r_1*sin((m*theta)+phi))
                A[i,j]=(1/(0.398942280*(2*pi)**0.5))*exp(-(R-r_0)**2/(2))
    return A
```

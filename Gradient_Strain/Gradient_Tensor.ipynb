### 1. Hallar el vector promedio

```python
#Esta imagen fue obtenida del libro Structural Geology de Allmendinger et al., (2012)
Image(filename="long.png")
```

La idea es decomponer los vectores (que a diferencia de la imagen solo están en 2D, el vector v esta en 3D) de desplazamiento en los valores de cos(alpha)y cos(betha) para poder luego sumarlos y hallar un vector general.

```python
#Acá se obtienen los vectores de desplazamiento en Norte y Este.
EastDisp = Data[:,2]
NorthDisp = Data[:,4]

#Acá se crea un archivo con las nuevas coordenadas

Datacordinates = open("NewCordinates.txt","w")
Datacordinates.write(" Easthing coordinate(m)   Northing coordinate(m)   U(E)    U(N) \n \n")
for i in range (size):
    Datacordinates.write("%s %s %s %s\n\n"%(Easting[i],Northing[i], EastDisp[i], NorthDisp[i]))
Datacordinates.close()
```

```python
#SE CREAN LISTAS PARA GUARDAR LA INFORMACIÓN
cos_alpha = zeros(size)
cos_betha = zeros(size)
sum_alpha = 0.0
sum_betha = 0.0

#la tabla de datos nos proporciona los valores de las proyecciones en el eje N y E para cada vector de desplazamiento
#para conocer el valor de los cosenos solo hace falta usar el vector de desplazamiento (Norte y Este por aparte) y 
#dividirlo sobre la norma de este

for i in range (0,size):
    cos_betha[i] = EastDisp[i]/(sqrt((EastDisp[i])**2 + (NorthDisp[i])**2))
    cos_alpha[i] = NorthDisp[i]/(sqrt((EastDisp[i])**2 + (NorthDisp[i])**2))
    #acá es necesario ir contando los valores y sumarlos
    sum_alpha = sum_alpha + cos_alpha[i]
    sum_betha = sum_betha + cos_betha[i]

print "el valor de cos(alpha) y cos(betha) es", sum_alpha, sum_betha

#para hallar la magnitud del nuevo vector se eleva al cuadrado los resultados anteriores
sum_alpha_2 = sum_alpha**2
sum_betha_2 = sum_betha**2

r = sqrt(sum_alpha_2+sum_betha_2)

print "la magnitud del nuevo vector es", r

#este parámetro muestra que tan correlacionados se encuentran los vectores que se asociaron al nuevo vector promedio.
#un valor cercano a 1 indica una alta correlación 
norm_r = r/(size-1)

print "el parámetro de correlación es", norm_r

#Se normalizan las componentes del nuevo vector, para que sea un vector unitario

final_l1 = (sum_alpha/r)**2
final_l2 = (sum_betha/r)**2

print "la mangnitud del vector normalizado es", sqrt(final_l1+final_l2)

angle = arctan (sqrt(final_l1)/sqrt(final_l2))
#esto se hace porque las componentes estan en los ejes negativos
anglecorrect = angle + pi

print "el ángulo gamma de este nuevo vector es:", anglecorrect
```

### 2. Hallar la matriz de cambio de coordenadas

Se define una matriz de transformación de dimensión 2 necesaria para un sistema de coordenadas donde el eje x quede paralelo al vector obtenido en el paso anterior. La matriz necesaria para esto es una matriz que permita un cambio de coordenadas con rotación en el eje de origen.

```python
#esta imagen fue obtenida de http://es.wikipedia.org
Image(filename="cambiodecoordenadas.png")
```

La matriz necesaria para esto es:

$$a_{ij} = \begin{bmatrix}
cos(\gamma) & -sin(\gamma) \\
sin(\gamma) & cos(\gamma)
\end{bmatrix}$$

donde gamma es el ángulo obtenido al final del paso anterior. Con esta matriz se tranformarán todas las coordenadas.

### 3. Transformar las coordenadas de GPS al nuevo sistema de coordenadas

```python
NewEast= zeros(size)
NewNorth= zeros(size)

for i in range (0,size):
    NewEast[i] = Easting[i]*cos(anglecorrect) - Northing[i]*sin(anglecorrect)
    NewNorth[i] = Easting[i]*sin(anglecorrect) + Northing[i]*cos(anglecorrect)
```

### 6. ¿Cómo lo vamos a dividir?

En el libro de Allmendinger et al., (2012) se propone un método muy complejo que no se usará acá, este método se llama triangulación de Delaunay (imagen de abajo). Acá vamos a intentar hacer algo parecido, simplemente para cada estación se hallarán las 3 estaciones más cercanas y con estas se estimará el tensor de deformación para esa zona. Se hará una simplificación, no se tomarán las barras de error para calcular el fit.

```python
#Esta imagen fue obtenida del libro Structural Geology de Allmendinger et al., (2012)

Image(filename="triangulacion.png")
```

### 7. Hallar los tensores para cada estación, los autovalores y autovectores correspondientes.

```python
#función que realiza un fit a partir de una regresión lineal, con esto se calcula la pendiente de la recta que viene siendo
#una componente del tensor de deformación e.
def e(x,y):
    slope = polyfit(x,y,1)
    return slope[0]
```

```python
listatensores = []
eigenvalores1 = []
eigenvectores1 = []
eigenvalores2 = []
eigenvectores2 = []
for i in range (0,size):
    listp = []
    listxc = zeros(4)
    listyc = zeros(4)
    listxu = zeros(4)
    listyu = zeros(4)
    for j in range (0,size):
        #se estima la distancia entre la estación y las otras estaciones
        r = sqrt((NewEast[i]-NewEast[j])**2 + (NewNorth[i]-NewNorth[j])**2)
        #se agregan estas distancias a una lista
        listp.append(r)
    for m in range (4):
        #se organiza la lista y se guardan las posiciones en otra lista con este comando
        prueba = argsort(listp)
        #se usan las 3 distancias menores
        listxc[m]= NewEast[prueba[m]]
        listyc[m]= NewNorth[prueba[m]]
        listxu[m]= EastDisp[prueba[m]]
        listyu[m]= NorthDisp[prueba[m]]
    #se usa la función para hallar la pendiente de los datos y así hallar las componentes del tensor
    a = e(listxc,listxu)
    b = e(listxc,listyu)
    c = e(listyc,listxu)
    d = e(listyc,listyu)
    #se arma el tensor de deformación
    tensor = zeros([2,2])
    tensor[0,0] = a
    tensor[0,1] = b
    tensor[1,0] = c
    tensor[1,1] = d
    #se halla el tensor simétrico aplicando la definición dada por el profesor Jaime en clase.
    tensor = 0.5*(tensor + tensor.T)
    #se hallan los autovalores y autovectores del tensor
    eigenvalues, eigenvectors = eig(tensor)
    #se guardan todo esto en una lista.
    listatensores.append(tensor)
    eigenvalores1.append(eigenvalues[0])
    eigenvectores1.append(eigenvectors[0])
    eigenvalores2.append(eigenvalues[1])
    eigenvectores2.append(eigenvectors[1])
```

### 8. Guardar resultados en archivos

```python
#se crea varios archivos .txt que guardan los resultados para cada estación
DataTensors = open("Tensores.txt","w")
DataTensors.write("    Tensor     Easthing coordinate(m)   Northing coordinate(m)\n \n")
for i in range (size):
    DataTensors.write("%s %s %s\n\n"%(listatensores[i],Easting[i],Northing[i]))
DataTensors.close()
DataVectors = open("Autovectores.txt","w")
DataVectors.write("    Autovector 1   Autovector 2    Easthing coordinate(m)   Northing coordinate(m)\n \n")
for i in range (size):
    DataVectors.write("%s %s %s %s\n\n"%(eigenvectores1[i],eigenvectores2[i],Easting[i],Northing[i]))
DataVectors.close()
DataValues = open("Autovalores.txt","w")
DataValues.write("    Autovalor 1  Autovalor 2     Easthing coordinate(m)   Northing coordinate(m)\n \n")
for i in range (size):
    DataValues.write("%s %s %s %s\n\n"%(eigenvalores1[i],eigenvalores2[i],Easting[i],Northing[i]))
DataValues.close()
```

### ¿Qué significa todo esto?

#### “As a structural geologist, I don’t believe in stress.” (John Ramsay)

Deformación vs Esfuerzo (Strain vs Stress)

Nosotros nunca vemos realmente el esfuerzo ni podemos medirlo directamente, ya que este es un campo tensorial. Lo único que se puede observar es el producto final de la acción de un esfuerzo sobre un material, esto se conoce como deformación. En nuestro ejemplo, lo que observamos es la deformación que sufrió Antofagasta, la cual fue medida con GPS. A partir de esto, buscamos inferir el tensor de deformación que sufrió la zona.

Realizar lo anterior, trae consigo muchas complicaciones a la hora de procesar los datos, por ejemplo saber cuantas estaciones se deben usar para los calculos, el problema de que las estaciones no estan uniformemente distribuidas, entre otros factores que nos impiden determinar exactamente el tensor de deformación.

Los dos autovalores obtenidos corresponden a las dos magnitudes de las deformaciones principales y sus correspondientes autovectores dan la orientación de los principales ejes de deformación.

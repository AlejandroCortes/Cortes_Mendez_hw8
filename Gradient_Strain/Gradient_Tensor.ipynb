# Tensor de deformación después de un terremoto

El uso de estaciones GPS (Global Position System) ha permitido a los geocientíficos detectar y medir pequeños movimientos de la tierra en tiempo real. En este ejercicio, se usarán los vectores de desplazamiento detectados y medidos por 32 estaciones de GPS en Antofagasta, luego del sismo que ocurrió en esta región en 1995 con Mw.(magnitud de momento) de 8.1

El siguiente notebook estima la parte simétrica del tensor de deformación en areas donde estan ubicadas las estaciones de GPS, y gráfica la magnitud del autovalor principal y la dirección del autovector principal correspondiente.

### Cargar datos de las estaciones

#### 1. Hallar el vector promedio

```python
Image(filename="long.png")
```

La idea es decomponer los vectores (que a diferencia de la imagen solo están en 2D, el vector v esta en 3D) de desplazamiento en los valores de cos(alpha)y cos(betha) para poder luego sumarlos y hallar un vector general.

```python
#Acá se obtienen los vectores de desplazamiento en Norte y Este.
EastDisp = Data[:,2]
NorthDisp = Data[:,4]
print NorthDisp
```

```python
#SE CREAN LISTAS PARA GUARDAR LA INFORMACIÓN
cos_alpha = zeros(size)
cos_betha = zeros(size)
sum_alpha = 0.0
sum_betha = 0.0

#la tabla de datos nos proporciona los valores de las proyecciones en el eje N y E para cada vector de desplazamiento
#para conocer el valor de los cosenos solo hace falta usar el vector de desplazamiento (Norte y Este por aparte) y 
#dividirlo sobre la norma de este

for i in range (1,size):
    cos_betha[i] = EastDisp[i]/(sqrt((EastDisp[i])**2 + (NorthDisp[i])**2))
    cos_alpha[i] = NorthDisp[i]/(sqrt((EastDisp[i])**2 + (NorthDisp[i])**2))
    #acá es necesario ir contando los valores y sumarlos
    sum_alpha = sum_alpha + cos_alpha[i]
    sum_betha = sum_betha + cos_betha[i]

print "el valor de cos(alpha) y cos(betha) es", sum_alpha, sum_betha

#para hallar la magnitud del nuevo vector se eleva al cuadrado los resultados anteriores
sum_alpha_2 = sum_alpha**2
sum_betha_2 = sum_betha**2

r = sqrt(sum_alpha_2+sum_betha_2)

print "la magnitud del nuevo vector es", r

#este parámetro muestra que tan correlacionados se encuentran los vectores que se asociaron al nuevo vector promedio.
#un valor cercano a 1 indica una alta correlación 
norm_r = r/(size-1)

print "el parámetro de correlación es", norm_r

#Se normalizan las componentes del nuevo vector, para que sea un vector unitario

final_l1 = (sum_alpha/r)**2
final_l2 = (sum_betha/r)**2

print "la mangnitud del vector normalizado es", sqrt(final_l1+final_l2)

angle = arctan (sqrt(final_l1)/sqrt(final_l2))
#esto se hace porque las componentes estan en los ejes negativos
anglecorrect = angle + pi

print "el ángulo gamma de este nuevo vector es:", anglecorrect
```

#### 2. Hallar la matriz de cambio de coordenadas

Se define una matriz de transformación de dimensión 2 necesaria para un sistema de coordenadas donde el eje x quede paralelo al vector obtenido en el paso anterior. La matriz necesaria para esto es una matriz que permita un cambio de coordenadas con rotación en el eje de origen.

```python
#esta imagen fue obtenida de http://es.wikipedia.org
Image(filename="cambiodecoordenadas.png")
```

La matriz necesaria para esto es:

$$a_{ij} = \begin{bmatrix}
cos(\gamma) & -sin(\gamma) \\
sin(\gamma) & cos(\gamma)
\end{bmatrix}$$

donde gamma es el ángulo obtenido al final del paso anterior. Con esta matriz se tranformarán todas las coordenadas.

#### 3. Transformar las coordenadas de GPS al nuevo sistema de coordenadas

```python
NewEast= zeros(size)
NewNorth= zeros(size)

for i in range (1,size):
    NewEast[i] = Easting[i]*cos(anglecorrect) - Northing[i]*sin(anglecorrect)
    NewNorth[i] = Easting[i]*sin(anglecorrect) + Northing[i]*cos(anglecorrect)
```
